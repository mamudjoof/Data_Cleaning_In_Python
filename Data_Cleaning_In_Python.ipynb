# Data Cleaning Portofolio

```python
import pandas as pd
```

```python
df=pd.read_csv('CriketTestMatchData.csv')
df
```

```python
 # Rename some of the colunms
```

```python
df1=df.rename(columns={'Inns':'Innings_Batted', 'NO':'Not_Outs', 'HS':'Highest_Inns_Score', 'BF':'Balls_Faced', 'SR':'Batting_Strike_Rate'})
```

```python
df1
```

```python
# Look out for the NULL values and fill them
```

```python
df1.isnull().any()
```

```python
df1[df1['Balls_Faced'].isna()==1]
```

```python
df1['Balls_Faced']=df1['Balls_Faced'].fillna(0)
```

```python
df1['Batting_Strike_Rate']=df1['Batting_Strike_Rate'].fillna(0)
```

```python
df1[df1['Player']=='ED Weekes (WI)']
```

```python
df1[df1['Player']=='CL Walcott (WI))']
```

```python
df1
```

```python
df1.isnull().any()
```

```python
# Drop Duplicates
```

```python
df1.drop_duplicates()
```

```python
#Separate the Span into two columns Start Date and End Date
```

```python
df1['Span']
```

```python
df1['Start_Date']=df1['Span'].str.split('-').str[0]
```

```python
df1['End_Date']=df1['Span'].str.split('-').str[1]
```

```python
df1
```

```python
#Separate Country name from the Player
```

```python
df1['Country']=df1['Player'].str.split('(').str[1]
```

```python
df1['Country']=df1['Country'].str.split(')').str[0]
```

```python
df1['Player']=df1['Player'].str.split('(').str[0]
```

```python
df1
```

```python
df1=df1.drop(columns=['Span'])
```

```python
df1
```

```python
#Change DataTypes
```

```python
df1.dtypes
```

```python
df1['Start_Date']=df1['Start_Date'].astype('int')
```

```python
df1['End_Date']=df1['End_Date'].astype('int')
```

```python
df1.dtypes
```

```python
df1['Highest_Inns_Score']=df1['Highest_Inns_Score'].str.split('*').str[0]
```

```python
df1['Highest_Inns_Score']=df1['Highest_Inns_Score'].astype('int')
```

```python
df1
```

```python
df1['Balls_Faced']=df1['Balls_Faced'].str.split('+').str[0]
```

```python
# Finally since the data is clean we can run some calculations
```

```python
# QUESTION ONE: Build Out Career_length Colunm
```

```python
df1['Career_length']=df1['End_Date']-df1['Start_Date']
```

```python
df1
```

```python
# Cricketers in this DF what is the Average Career Length
```

```python
df1['Career_length'].mean()
```

```python
# AVG Batting_rate for Cricketers who played over 10 years.
```

```python
df1[df1['Career_length']>10]['Batting_Strike_Rate'].mean()
```

```python
# Find the number of cricketers who played before 1960.
```

```python
df1[df1['Start_Date']<1960]['Player'].count()
```

```python
# Max Highest Inns Score by Country
```

```python
df1.groupby('Country')['Highest_Inns_Score'].max()
```
